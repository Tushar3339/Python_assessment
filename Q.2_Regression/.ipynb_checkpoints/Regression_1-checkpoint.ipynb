```python
# Create a correlation matrix
correlation_matrix = data.corr()

# Set the correlation threshold
threshold = 0.8

# Find groups of highly correlated columns
highly_correlated_groups = []
visited_columns = set()

for column in correlation_matrix.columns:
    if column not in visited_columns:
        correlated_group = set(correlation_matrix.index[(correlation_matrix[column] >= threshold)])
        highly_correlated_groups.append(correlated_group)
        visited_columns.update(correlated_group)

# Print the highly correlated groups
for i, group in enumerate(highly_correlated_groups):
    print(f"Group {i+1}: {', '.join(group)}")
```

### There are no big groups of highly correlated groups

```python
Q1 = data['shares'].quantile(0.25)
Q3 = data['shares'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
data_filtered = data[(data['shares'] >= lower_bound) & (data['shares'] <= upper_bound)]
```

```python
data_filtered.shape
```

## Tried taking iteratively from 20-57 highly correlated columns with output
## This gave max R2 score 0.039 with 50 columns selected

```python
CorrMat = data.corr()
for i in range(20,57,1):
    corr_shares = CorrMat['shares'].sort_values(ascending=False).head(i+1)[1:]
    columns_selected = corr_shares.index
    data_croped = data_filtered.copy()
    data_croped = pd.concat([data_croped[columns_selected], data_croped['shares']], axis = 1)
    
    X = data_croped.iloc[:, :-1].values
    y = data_croped.iloc[:, -1].values
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
    
    regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
    regressor.fit(X_train, y_train)
    y_pred = regressor.predict(X_test)
    
    print(f' for i = {i}: {r2_score(y_test, y_pred)}')
```

### 0.039 is best score

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X_train_n = sc_X.fit_transform(X_train_n)
y_train_n = sc_y.fit_transform(y_train_n)
```

```python
y_pred_n = sc_y.inverse_transform(regressor.predict(sc_X.transform(X_test_n)).reshape(-1,1))
```

```python

r2_score(y_test_n, y_pred_n)
```

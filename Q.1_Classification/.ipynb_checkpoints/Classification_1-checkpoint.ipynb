```python
CorrMat = data.corr()
plt.figure(figsize = (20,20))
sns.heatmap(CorrMat, annot=True)
```

```python
# Create a correlation matrix
correlation_matrix = data.corr()

# Set the correlation threshold
threshold = 0.8

# Find groups of highly correlated columns
highly_correlated_groups = []
visited_columns = set()

for column in correlation_matrix.columns:
    if column not in visited_columns:
        correlated_group = set(correlation_matrix.index[(correlation_matrix[column] >= threshold)])
        highly_correlated_groups.append(correlated_group)
        visited_columns.update(correlated_group)

# Print the highly correlated groups
for i, group in enumerate(highly_correlated_groups):
    print(f"Group {i+1}: {', '.join(group)}")
```

#### BILL_AMT6, BILL_AMT5, BILL_AMT3, BILL_AMT4, BILL_AMT2, BILL_AMT1 are highly correlated. We will keep only one

```python
columns_to_drop = ['BILL_AMT5', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT2', 'BILL_AMT1']
data.rename(columns={'BILL_AMT6': 'BILL_AMT'}, inplace=True)
reduced_data = data.drop(columns_to_drop, axis=1)
original_data = data

reduced_data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import plot_confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC, NuSVC, LinearSVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb
```

```python
def train_model_helper(model, X, y, test_size=0.25, random_state=0, plot_confusion_matrix = False, Model_Name=''):
    
    if Model_Name=='':
        Model_Name = model.__class__.__name__
    print(f"Model name: {Model_Name} \nTest Size:{test_size}")
    
    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    
    # Feature scaling
    sc = StandardScaler()
    X_train = sc.fit_transform(X_train)
    X_test = sc.transform(X_test)
    
    # Train the model
    model.fit(X_train, y_train)
    
    # Make predictions on the test set
    y_pred = model.predict(X_test)
    
    if plot_confusion_matrix:
        # Print confusion matrix
        cm = confusion_matrix(y_test, y_pred)
        plt.figure(figsize=(6, 6))
        sns.heatmap(cm, annot=True, cmap='Blues', fmt='g')
        plt.title(f"{Model_Name} Confusion Matrix")
        plt.xlabel("Predicted Label")
        plt.ylabel("True Label")
        plt.show()
    
    
    # Print accuracy score
    accuracy = round(accuracy_score(y_test, y_pred)*100, 2)
    print(f"{Model_Name} Accuracy: {accuracy:.2f}%")
    
    return accuracy
```

```python
models_acc = {}
data = [original_data, reduced_data]
data_names = ['original_data', 'reduced_data']
Test_Data_Sizes = [0.15, 0.20, 0.25, 0.30]
```

```python
def train_model(model_name, model, is_plot_matrix = False, test_data_size=[0.25]):
    models_acc[model_name] = []
    
    for size in test_data_size:
        for index, curr_data in enumerate(data):
            #splitting data in input and output
            X = curr_data.iloc[:, :-1].values
            y = curr_data.iloc[:, -1].values

            #training the model for given data
            print(f'\n\nFor {data_names[index]}: \n')
            acc = train_model_helper(model, X, y, Model_Name=model_name, test_size=size, plot_confusion_matrix=is_plot_matrix)

            #appending the accuracy to dictionary
            models_acc[model_name].append(acc)
    
```

```python
models = {
    'Logistic Regression':LogisticRegression(random_state=0),
    'KNN':KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2),
    'SVC_Linear':SVC(kernel = 'linear', random_state = 0),
    'SVC_Radial':SVC(kernel = 'rbf', random_state = 0),
    'Naive_Bayes':GaussianNB(),
    'Decision_Tree':DecisionTreeClassifier(criterion = 'entropy', random_state = 0),
    'Random_Forest':RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0),
    'XG_Boost':xgb.XGBClassifier()
}
```

```python
for name,model in models.items():
    train_model(name, model, test_data_size=Test_Data_Sizes, is_plot_matrix = True)
    
```

```python
results = pd.DataFrame(models_acc)
results = results.transpose()
results.columns = ['acc_org_15', 'acc_red_15', 'acc_org_20', 'acc_red_20', 'acc_org_25', 'acc_red_25', 'acc_org_30', 'acc_red_30']
results
```

```python
print(results.max())
print(results.idxmax())
```

```python
results.max()
```

### In all testsizes SVC with radial kernel is performing best 

```python
ax = results.plot(kind='bar', figsize=(16, 6))
ax.set_ylabel('Accuracy')
ax.set_xlabel('Model and Dataset Size')
plt.title('Model Accuracy for Different Dataset Sizes')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Dataset Size', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

## 1) SVC Radial and XGBoost perform consistently well across different test sizes.
## 2) Reduced data generally shows similar or slightly better accuracy compared to the original data for most models.
## 3) Naive Bayes shows significant variations in accuracy, possibly influenced by the data reduction process.

### In most cases, even with 5 reduced columns, the accuracy of the models remains the same or improves slightly. This indicates that the data reduction process successfully captures the important information while removing unnecessary correlations, resulting in a more efficient and accurate model.

## Using 'class_weights = balanced' for some models to deal with imbalance in data 

```python
models_acc2 = {}
data = [original_data, reduced_data]
data_names = ['original_data', 'reduced_data']
Test_Data_Sizes = [0.15, 0.20, 0.25, 0.30]

models2 = {
    'Logistic Regression':LogisticRegression(random_state=0, class_weight='balanced'),
    'SVC_Linear':SVC(kernel = 'linear', random_state = 0, class_weight='balanced'),
    'SVC_Radial':SVC(kernel = 'rbf', random_state = 0, class_weight='balanced'),
    'Random_Forest':RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0, class_weight='balanced')
}
```

```python
def train_model2(model_name, model, plot_confusion_matrix = False, test_data_size=[0.25]):
    models_acc2[model_name] = []
    
    for size in test_data_size:
        for index, curr_data in enumerate(data):
            #splitting data in input and output
            X = curr_data.iloc[:, :-1].values
            y = curr_data.iloc[:, -1].values

            #training the model for given data
            print(f'\n\nFor {data_names[index]}: \n')
            acc = train_model_helper(model, X, y, Model_Name=model_name, test_size=size)

            #appending the accuracy to dictionary
            models_acc2[model_name].append(acc)
```

```python
for name,model in models2.items():
    train_model2(name, model, test_data_size=Test_Data_Sizes)
```
